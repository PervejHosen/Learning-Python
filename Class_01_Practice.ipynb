# Python code snippet

**A snippet of Python to make a visualization**

```python
#The code cell below uses numpu to generate some random data, and uses matplotlib to visualize it
import numpy as np
from matplotlib import pyplot as plt

y = 200+np.random.randn(100)
x = [x for x in range(len(y))]

plt.plot(x, y, '-')
plt.fill_between(x, y, 195, where=(y>195), facecolor='g', alpha=0.6)

plt.title('Sample Visualization')
plt.show()
```

##**Lists**:
* Lists are used to store multiple items in a single variable.
* Lists are one of 4 built-in data types in Python used to store collections of data
* the other 3 are Tuple, Set, and Dictionary, all with different qualities and usage.
* Lists are created using square brackets:

```python
my_list = [1, 2, '3', True]     # We assume this list won't mutate for each example below

print(my_list)
print(len(my_list))             #4
print(my_list.index('3'))       #2
print(my_list.count(2))         # 1 --> count how many times 2 appears
print(my_list[3])               # True
print(my_list[1:])              # [2, '3', True]
print(my_list[:1])              # [1]
print(my_list[-1])              # True
print(my_list[::1])             # [1, 2, '3', True]
print(my_list[::-1])            # [True, '3', 2, 1]
print(my_list[0:3:2])           # [1, '3']

# : is called slicing and has the format [ start : end : step ]
```

```python
# Add to List
my_list * 2
```

```python
my_list + [100] # --> doesn't mutate original list, creates new one
```

```python
my_list.append(100) #Mutate original list
print(my_list)
```

```python
my_list.extend([100, 200])
print(my_list)
```

```python
my_list.insert(2, '!!!')  #[1,2,'!!!', '3', True] --> Inserts item at index and moves the rest to the right
print(my_list)
```

```python
' '.join(['Hello','There']) #Joins elements using string as separator
```

```python
# Copy a List
basket = ['apples', 'pears', 'oranges']
new_basket = basket.copy()
print(new_basket)
```

```python
# clearing a list
x = [1,2,3]
x.clear()
print(x)
```

```python
# Ordering list
x = [1,2,5,3]
x.sort()
print(x)
```

```python
# Useful operations
print(1 in [1,2,5,3])     # True
print(min([1,2,3,4,5]))   # 1
print(max([1,2,3,4,5]))   # 5
sum([1,2,3,4,5])          # 15
```

```python
# Get First and Last element of a list
mList = [63, 21, 30, 14, 35, 26, 77, 18, 49, 10]
first, *x, last = mList
print(first)                #63
print(last)                 #10
```

##**Dictionaries**:
* Dictionaries are used to store data values in key:value pairs.
* Also known as mappings or hash tables.
* A dictionary is a collection which is ordered.
* Changeable and do not allow duplicates.
* They are key value pairs that are guaranteed to retain order of insertion.
* Dictionaries are written with curly brackets, and have keys and values:

```python
my_dict = {
    'name': 'Pervej',
    'age': 25,
    'magic_power': False}
print(my_dict)
```

```python
my_dict1 = {
    'name': ['Pervej',  'Hosen', 'Rasel'],
    'age': [24, 25, 26],
    'magic_power': [False, True, True]}
print(my_dict1)
```

```python
print(my_dict1['name'])
print(len(my_dict1))
print(list(my_dict.keys()))
print(list(my_dict.values()))
print(list(my_dict1.keys()))
print(list(my_dict1.values()))
print(list(my_dict.items()))
print(list(my_dict1.items()))

print(my_dict.get('age'))
print(my_dict1.get('age'))

print(my_dict.get('ages', 0 ))        # 0 --> Returns default (2nd param) if key is not found
```

```python
#Remove key
del my_dict['name']

print(my_dict)
```

```python
#Update key
my_dict.update({'cool': True})
print(my_dict)
```

```python
#Create a dict From two collection
new_dict = dict(zip(['name','age','magic_power'],['Andrei',32, False]))
new_dict
```

##**Tuples:**
* Tuples are used to store multiple items in a single variable.
* A tuple is a collection which is ordered and unchangeable.
* Tuples are written with round brackets.

```python
#creating tuples
my_tuple = ('apple','grapes','mango', 'grapes')
print(my_tuple)
```

```python
print(len(my_tuple))      # 4
print(my_tuple[2])        #mango
print(my_tuple[-1])       #grapes
```

```python
# Immutability
my_tuple[1] = 'donuts'  # TypeError
```

```python
my_tuple.append('candy')# AttributeError
```

```python
# Methods
my_tuple.index('grapes')      # 1
```

```python
# Methods
my_tuple.count('grapes')      # 2
```

```python
# Zip
list(zip([1,2,3], [4,5,6]))   # [(1, 4), (2, 5), (3, 6)]
```

```python
# unzip
z = [(1, 2), (3, 4), (5, 6), (7, 8)]    # Some output of zip() function
unzip = lambda z: list(zip(*z))
unzip(z)
```

##**Sets:**
* Sets are used to store multiple items in a single variable.
* Set is one of 4 built-in data types in Python used to store collections of data, the other 3 are List, Tuple, and Dictionary, all with different qualities and usage.
* A set is a collection which is unordered, unchangeable*, and unindexed.

```python
my_set = set()
my_set.add(1)           # {1}
my_set.add(100)         # {1, 100}
my_set.add(100)         # {1, 100} --> no duplicates!
print(my_set)
```

```python
new_list = [1,2,3,3,3,4,4,5,6,1]
set(new_list)                       # {1, 2, 3, 4, 5, 6}
```

```python
my_set.remove(100)                  #Raises KeyError if element not found
```

```python
my_set.discard(100)                 #Doesn't Raise KeyError if element not found
```

```python
my_set.clear()
```

```python
set1 = {1,2,3}
set2 = {3,4,8}
set3 = set1.union(set2)
print(set3)
set4 = set1.intersection(set2)
print(set4)
```
