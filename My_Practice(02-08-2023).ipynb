# Python code snippet

**A snippet of Python to make a visualization**

```python
#The code cell below uses numpu to generate some random data, and uses matplotlib to visualize it
import numpy as np
from matplotlib import pyplot as plt

y = 200+np.random.randn(100)
x = [x for x in range(len(y))]

plt.plot(x, y, '-')
plt.fill_between(x, y, 195, where=(y>195), facecolor='g', alpha=0.6)

plt.title('Sample Visualization')
plt.show()
```

```python
#to check the length of the assign variables
print(len(x))
print(len(y))
```

#Types of data/variables in Python

**Types**: Numbers, Strings, Boolean, Lists, Dictionaries, Tuples, Sets

##**Numbers:**
Python has two main types of numbers.
* int [integers]
* float [floating point numbers or decimals]

**lets check some data types**

```python
print(type(1))      # int
print(type(-10))    # int
print(type(0))      # int
print(type(0.0))    # float
print(type(2.2))    # float
print(type(4E2))    # float - 4*10 to the power of 2
```

##**Strings:**
* String is a collection of alphabets, words or other characters.
* Stored as sequences of letters in memory.
* Surrounded by either single quotation marks, or double quotation marks.

```python
print('Hello Pervej, Happy journey on learning Python, Have fun.')
print(len('Hey Pervej Hosen'))
type("Pervej")
```

**Assigning string to variable**

```python
#Assigning a string to a variable

a = "Hello Pervej"
print(a)
```

```python
#Assigning multiline string to a variable
#we can use three double quotes

a = """Hey Pervej, hope you are doing well.
You just started learning this increadible programing language 'Python'.
Happy journey on learning thid language.
Have fun."""
print(a)
```

**Slicing Strings**

```python
name = 'Pervej Hosen'
print(name[3])
```

```python
#Slice From the Start
name = 'Pervej Hosen'
print(name[:6])
```

```python
#Slice To the End
name = 'Pervej Hosen'
print(name[6:])
```

**Modify Strings**

```python
#uppper Case
a = "Hey Pervej"
print(a.upper())
```

```python
#Lower Case
a = "Hey Pervej"
print(a.lower())
```

```python
#Remove Whitespace
a = " Hello, World! "
print(a.strip()) # returns "Hello, World!"
```

```python
#Replace String
a = "Hello, World!"
print(a.replace("H", "J"))
```

```python
#Split String
a = "Hello, World!"
print(a.split(",")) # returns ['Hello', ' World!']
```

**Concatenate String**

```python
a = "Hello"
b = "World"
c = a + b
print(c)
```

**String Formatting**

```python
name1 = 'Pervej'
name2 = 'Hosen'
print(f'Hello there {name1} {name2}')
```

```python
#summary
name = 'Pervej Hosen'
print(name[3])
print(name[1:])
print(name[:1])
print(name[::-1])
print(name[0:10:2]) #: is called slicing and has the format [ start : end : step ]

print('Hi there ' + 'Rossi') # 'Hi there Rossi' --> This is called string concatenation

print(' I am alone '.strip()) #Strips all whitespace

print('ok, I am done.'.capitalize())

print('On an island'.strip('d'))

print('Help me'.replace('me', 'you'))

print('but life is good!'.split())

print('oh hi there'.find('i'))
```
